```python
# calculating sum,mean,mode of a particular column 

sum_data =  df["sepal_length"].sum()
mean_data = df["sepal_length"].mean()
mode_data = df["sepal_length"].mode()
print("sum is ::",sum_data)
print("mean is::" , mean_data)
print("mode is ::",mode_data)
```

```python
# calculating max , min

max_data =  df["sepal_width"].max()
print("maximum value is::" ,max_data)

min_data = df["sepal_width"].min()
print("minimum value is ::",min_data)
```

```python

import seaborn as sns
import pandas as pd

# Load the iris dataset
iris = sns.load_dataset("iris")

# Drop the 'species' column before calculating correlation, as it's categorical
iris_numeric = iris.drop(columns=['species'])

# Calculate the correlation matrix for the numeric columns only
correlation_matrix = iris_numeric.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, cmap="YlGnBu", linecolor='white', linewidths=1)
```
